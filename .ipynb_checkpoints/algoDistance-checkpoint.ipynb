```python
#creat position dataframe from the input jsondata

def posData(xty, x, yty, y, xminOcc = 0, yminOcc= 0):
    """return positions data after normalization (in [0,1])"""
    jsonData, _, _, _ = tsv2json(xty =xty, x=x , xminocc = xminOcc,yty = yty,y = y, yminocc = yminOcc, verbose = False)
    
    N = len(jsonData)
    dataDict = {}
    for la in jsonData:
        assert(len(la.get('data'))==1)
        data = la.get('data')[0]
        dataDict[data.get('label')]={'x': data.get('x'), 'y': data.get('y')}
        
    XY = pd.DataFrame(data=dataDict).T
    #norm = np.linalg.norm(XY.values, axis = 0)
    return (XY-XY.min())/(XY.max()-XY.min())
```

```python
def dist(p1,p2):
    return np.sum((p1-p2)**2)**0.5
```

```python
def langInter(posG1,posG2):
    """
    extract languges in both graph1 and graph2
    posG1, posG2 : dataframes of languges positions in graph1 & graph2
    """
    lanG1 = posG1.index.tolist()
    lanG2 = posG2.index.tolist()
    print(lanG1 ==lanG2)
    
    if lanG1 !=lanG2:
        print("extract languages")
        langN = [lan for lan in lanG1 if lan in lanG2 ]
        #extract positions of langN 
        posG1 =  posG1.loc[langN]
        posG2 = posG2.loc[langN]
        assert(posG1.index.tolist() == posG2.index.tolist())
    return posG1, posG2
```

```python
posG1, posG2 = langInter(posG1,posG2)
```

```python
def distGraphDep(g1,g2):
    """
    languages' deplacement between g1 and g2 
    g1,g2: 2 dataframes contain positions of languages in both graph1 and graph2
    """
    g1,g2 = langInter(g1,g2)
    distSum = 0
    for la in g1.index:
        dl = dist(g1.loc[la].values, g2.loc[la].values)
        distSum += dl
        #print( la, "\ndeplacement: ", dl)
        #distSum += dist(g1.loc[la].values, g2.loc[la].values)
    return distSum
```

```python
#try with distance, xminocc = 0, yminocc = 0

typ = ['distance']

testGr = [(ty,opt) for ty in typ for opt in getoptions(ty)]
print(len(testGr))

distData = {'deplacement':{},
            'cloudForm':{}}
mappings = {}
```

```python
('distance','subj:obj') in testGr
```

```python
#try with distance, xminocc = 0, yminocc = 0

for g1 in testGr:
    k1 = ':'.join(g1)
    distData['deplacement'][k1] = distData['deplacement'].get(k1,{})
    distData['cloudForm'][k1] = distData['cloudForm'].get(k1,{})
    mappings[k1]={}
    
    for g2 in testGr:
        
        k2 = ':'.join(g2)
        distData['deplacement'][k2] = distData['deplacement'].get(k2,{})
        distData['cloudForm'][k2] = distData['cloudForm'].get(k2,{})
        mappings[k2]=mappings.get(k2,{})
        
        if distData['deplacement'][k1].get(k2)==None:
            #jsonG1, nblang1, _, _= tsv2json(xty = g1[0],x=g1[1],xminocc = 0,yty=g1[0],y=g1[1],yminocc = 0)
            #jsonG2, nblang2, _, _= tsv2json(xty = g2[0],x=g2[1],xminocc = 0,yty=g2[0],y=g2[1],yminocc = 0)
            posG1 = posData(g1[0],g1[1],g1[0],g1[1])
            posG2 = posData(g2[0],g2[1],g2[0],g2[1])

            distDep = distGraphDep(posG1,posG2)
            matches, distDTW, mapping1, mapping2, matrix = simpledtw.dtw(posG1.values, posG2.values)
            distData['deplacement'][k1][k2]= distDep
            distData['deplacement'][k2][k1]= distDep
            
            assert(distData['cloudForm'][k1].get(k2) == None)
            #print("--------\n", distData['cloudForm'][k1].get(k2))
            distData['cloudForm'][k1][k2]= distDTW
            distData['cloudForm'][k2][k1]= distDTW
            mappings[k1][k2] = mapping1
            mappings[k2][k1] = mapping2
```
